### 연속하는 정수의 합을 구하기 through 오름차순 정렬
#### a와 b 정수를 input받고 a부터b까지의 합 구하기

```python
#방법1
a,b = map(int,input('정수를 입력하십시오 :').split(','))

if a > b:          #★★★★★★★★★ 정렬 ★★★★★★★★★★★★★  a와 b 크기비교해서 큰게 b로 !!!!!!!!!!!!!
    a,b = b,a      #★★★★★★★★★★★★★★★★★★★★★★★★★ 두 개를 교환할 때는 단일문 사용!!
    
sum = 0
for i in range(a, b+1):
    sum += i
print(sum)
```

```python
# 방법2
a,b = map(int,input('정수를 입력하십시오 :').split(','))

if a > b:       
    a,b = b,a
sum = 0
for i in range(a, b):
    print(f'{i} +', end='')
    sum += i
print(f'{b} =', end='')
sum+=b

print(sum)
```

#### 줄바꿈 없이 +,-를 연속하여 출력해보자.

```python
n = int(input('몇 개를 출력하시겠습니까 ? : '))

for i in range(n):
    if i % 2:   # 홀수 인경우
        print('-',end='')
    else:
        print('+',end='')
```

```python
n = int(input('몇 개를 출력하시겠습니까 ? : '))
        
for i in range(1,n+1):
    if i % 2:
        print('+',end='')
    else:
        print('-',end='')
        
#  이 코드는 변수 i 를 0에서 n-1까지 1씩 증가시키는게 아닌 1부터 n까지 1씩 증가한다.
```

### 반복 과정에서 조건 판단하기
#### * 을 n개 출력하되 w개 마다 줄바꿈을 하는 프로그램

- 줄바꿈 : print() 
<<<단순하게>>>

```python
n = int(input('몇 개를 출력하시겠습니까 ? : '))
w = int(input('몇 개마다 줄바꿈을 하시겠습니까 ?: '))

for i in range(n):    # 총 n개를 출력
        print('*',end='')  # 한개씩 * 출력
        if i % w == w-1:  # 하다가 w가 w-1가 같아지면 (0부터 시작하므로)
            print()       # 줄 바꿈
            
if n % w:              
    print()            # 이 코드는 뭐지
```

### 위 코드:
- for문 반복 할때마다 if 수행하므로 비효율적
- 아래 코드로 개선

```python
n = int(input('몇 개를 출력하시겠습니까 ? : '))
w = int(input('몇 개마다 줄바꿈을 하시겠습니까 ?: '))

for _ in range(n//w):
    print('*'* w)
if n % w ==0:
    pass
else:
    print((n%w)*'*')
```

## 비교 연산자를 연속으로 사용하는 방법 / 드모르간 법칙

```python
# 2자리 양수 입력받기
while True:
    n = int(input('두 자리의 양수를 입력하십시오'))  # while 문 안에 (아래에) 있어야 한다...!!
    if 10 <= n and n <= 99:
        break
print('성공')
```

- 드모르간 법칙 사용

```python
# 2자리 양수 입력받기 by using 드모르간 법칙
while True:
    n = int(input('두 자리의 양수를 입력하십시오'))  # while 문 안에 (아래에) 있어야 한다...!!
    if not (n > 99  or  n < 10):
        break
print('성공')
```

# 구구단, 삼각형 그리기 

```python
# 구구단 곱셈표 출력하기
print("      ","구구단 표를 출력합니다.")
print('*' * 37)
for x in range(1,10):                    # 행
    for y in range(1,10):                # 열
        print(f"{x * y : 3}",end= ' ')   # ★★★  x * y:3 ==> x * y를 3자리로 가지런하게 출력  ★★★
    print()  # ★★ 행 변경!!!! ★★
print('*' * 37)
```

```python
# 직각 이등변 삼각형으로 출력하기 *모양으로!!!
# 왼쪽 아래가 직각임

n = int(input("짧은 변의 길이를 입력하십시오 : "))

for i in range(n):
    for j in range(i+1):
        print('*',end=' ')
    print()               # 줄 바꿈
```

```python
# ★★★★★★오른쪽 아래가 직각 일때....★★★★★★
n = int(input('짧은 변의 길이를 입력하십시오 : '))
for i in range(n):
    for _ in range(n-i-1):
        print(' ', end= ' ')
    for _ in range(i+1):
        print('*', end= ' ')
    print()
```
